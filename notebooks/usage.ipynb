# LoRA Fine-Tune usage

This notebook is intentionally thin: it just shows how to call the CLI.

## Setup
```bash
pip install -U transformers accelerate datasets peft bitsandbytes sentencepiece
export PYTHONPATH=$(pwd)/lora_finetune/src:$PYTHONPATH
```

## Train
```bash
python -m lora_finetune.main train --config ../configs/qwen3-4b.yaml
```

## Infer
```bash
python -m lora_finetune.main infer --config ../configs/qwen3-4b.yaml --prompt "Explain how tax exemption approval continues."
```

Change the YAML to adjust tokenizer, data paths, or LoRA hyperparams.
